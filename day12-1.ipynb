# Day 12: [Garden Groups](https://adventofcode.com/2024/day/12)

## Pseudo code
1. Use the garden plot (input) to identify unique number of characters, then store as dictionary
2. Calculate the area of each character 
    - Area = number of chracters in the plot 
3. Calculate the perimeter of each character
    - Perimeter : Based on presence of same character in adjecent locations
3. Calculate the (sum of area * perimeter)

## 1. Identify unique characters in the input

```python
with open ('day12-1.txt', 'r') as f:
    plot = [line.strip() for line in f]


print(f"Size of the plot : {len(plot)} (Width) X {len(plot[0])} (Height)")
plot_u = list(set("".join(plot)))
plot_u.sort()
print(f'Total number of unique characters in the plot : {len(plot_u)} ({plot_u})')
```

## 2. Calculate area and perimeter

### - Positions in matrix and list
| Starting point | (x,y) matrix | (i,j) list|
| --- | --- | --- |
| Up | (x,y+1) | (i-1,j) |
| Down | (x,y-1) | (i+1,j) |
| Left | (x-1,y) | (i,j-1) |
| Right | (x+1,y) | (i,j+1) |

### 2-1. Identify groups of adjacent locations

```python
# Store all locations of each character

plants = {c:[] for c in plot_u} # To store [area, perimeter] as values to a key

for i in range(len(plot)):
    for j in range(len(plot[0])):
        plants[plot[i][j]].append([i,j])
```

```python
# Group the locations into regions depending on the proximity
price = 0

for key, value in plants.items():
    locations = value.copy()
    regions = []
    for f in value:
        if any(f in g for g in regions):
            continue
        regions.append([f])
        locations.remove(f)
        for group in regions:
            for l in group:
                try:
                    presence = False
                    if ([l[0]-1,l[1]] in locations) and (l[0]-1 >= 0):  # Look upward
                        group.append([l[0]-1,l[1]])
                        locations.remove([l[0]-1,l[1]])
                    if ([l[0]+1,l[1]] in locations):                    # Look downward
                        group.append([l[0]+1,l[1]])
                        locations.remove([l[0]+1,l[1]])
                    if ([l[0],l[1]-1] in locations) and (l[1]-1 >= 0):  # Look left
                        group.append([l[0],l[1]-1])
                        locations.remove([l[0],l[1]-1])
                    if ([l[0],l[1]+1] in locations):                    # Look right
                        group.append([l[0],l[1]+1])
                        locations.remove([l[0],l[1]+1])
                except:
                    pass

# Calculate area and perimeter of each region
    for region in regions:
        counter = 1
        area = 0
        perim = 0
        for l in region:
            area += 1                           # Update area
            perim_temp = 4                           # To calculate perimeter for the given location
            try:                                # Up
                if ([l[0]-1,l[1]] in region) and (l[0]-1 >= 0):
                    perim_temp -= 1
            except:
                pass
            try:                                # Down
                if [l[0]+1,l[1]] in region:
                    perim_temp -= 1
            except:
                pass
            try:                                # Right
                if [l[0],l[1]+1] in region:
                    perim_temp -= 1
            except:
                pass
            try:                                # Left
                if ([l[0],l[1]-1] in region) and (l[1]-1 >= 0):
                    perim_temp -= 1
            except:
                pass
            perim += perim_temp
        price += area*perim
        print(f'{key} plants region {counter}: {area} (area) X {perim} (perimeter) = {area*perim}')
        counter += 1
print(f'Total price : {price}')
```
