# Day 13: [Claw Contraption](https://adventofcode.com/2024/day13)

## Pseudo code
1. Filter button configurations and prize location from the input text
2. Check if there are possible combinations for reaching the prize location
3. Calculate the tokens

## 1. Input to nested list 
Store each input as [Button A movement , Button B movement, Prize location]

```python
with open ('day13-1.txt', 'r') as f:
  input_text = [lines.strip() for lines in f]
```

```python
import re

inputs = []
filter = r"X[=+](\d+), Y[=+](\d+)"          # To find numbers after X+ and Y+
for i in range(0,len(input_text),4):
    temp = []                           # To store combinations temporarily
    temp.append([(int(i),int(j)) for i, j in re.findall(filter,input_text[i])][0])    # Button A
    temp.append([(int(i),int(j)) for i, j in re.findall(filter,input_text[i+1])][0])  # Button B
    temp.append([(int(i),int(j)) for i, j in re.findall(filter,input_text[i+2])][0])    # Button C
    inputs.append(temp)
```

## 2. Check for possible combinations

Try all possible combinations to find the case where sum of button A's X-axis and button B's X-axis equals with the prize's X-axis. When found, substitute to Button A's Y-axis and Button B's Y-axis to verify whether it also equals with prize's Y-axis. 

```python
tokens = 0
for input in inputs:
    a, b, p = input
    for i in range(p[0]//a[0]+1):
        for j in range(p[0]//b[0]+1):
            if (a[0]*i + b[0]*j == p[0]) and (a[1]*i + b[1]*j == p[1]):
                temp = i*3 + j*1
                print(f'Combination found for Prize {p}\n- Button A: {i} times\n- Button B: {j} times\n- Total: {temp} tokens')
                tokens += temp
                break
```

```python
print(f'Total number of tokens : {tokens}')
```
