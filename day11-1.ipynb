# Day 11: [Plutonian Pebbles](https://adventofcode.com/2024/day/11)

## Pseudo code

1. Store input as a list
2. Transform the input numbers with the below rules, for 25 times
-  0 : replace with 1
-  Even number of digit : split the number in half and store both left side and right side number separately
-  Others : multiply number by 2024
3. Calculate length of the list

## 1. Input to list

```python
with open('day11-1.txt', 'r') as f:
    stones = [line.strip().split() for line in f][0]
    stones = [int(n) for n in stones]
stones
```

## 2. Transform the numbers

```python
print(f'Inital list : {stones}')
for b in range(25):
    temp = []   # To temporarily store new stone order
    for i in range(len(stones)):
        if stones[i] == 0:
            temp.append(1)
        elif len(str(stones[i])) % 2 == 0:
            first_num = str(stones[i])[:len(str(stones[i])) // 2]
            temp.append(int(first_num))
            second_num = str(stones[i])[len(str(stones[i])) // 2:]
            temp.append(int(second_num))
        else:
            temp.append(stones[i]*2024)
    stones = temp
    print(f'Transformation {b+1} : {stones}')
```

```python
print(f'Number counts after {b+1}th transformation : {len(stones)}')
```
